```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
car_dataframe = pd.read_csv("data/car details v4.csv")
car_dataframe
```

```python
print(car_dataframe.shape)
print(car_dataframe.columns)

hot_encode = pd.get_dummies(car_dataframe)
print(hot_encode.shape)
pd.options.display.max_columns = 30000
for col in hot_encode.columns:
    print(col)
```

```python
x = (car_dataframe.loc[:, ["Fuel Tank Capacity"]]).to_numpy().reshape(-1).astype('float64')
y = (car_dataframe.loc[:, ["Price"]]).to_numpy().reshape(-1).astype('float64')

plt.plot(x, y, 'bo', ms=1) 

plt.show()
```
